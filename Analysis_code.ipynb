```python
import pandas as pd

import scipy
```

```python
stat1 = pd.read_csv('archive/stats1.csv')
pantic = pd.read_csv('archive/participants.csv')

model1 = pd.merge(stat1, pantic, on = 'id')

```

```python
model1
```

```python
model2 = model1[['win', 'firstblood','matchid']]

model2
```

```python
model3 = model2.groupby('matchid').first()

model3
```

```python
model4 = pd.crosstab(model3['firstblood'], model3['win'])
```

```python
scipy.stats.chi2_contingency(model4)
```

### 结论
通过卡方分析，计算结果p远小于0.001，所以我们拒绝H0假设，得出一血率与胜率呈现强相关性。
